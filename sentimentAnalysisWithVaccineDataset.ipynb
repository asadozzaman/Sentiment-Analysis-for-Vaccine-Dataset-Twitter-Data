### Sentiment analysis

**What is sentiment analysis?**

Sentiment analysis in NLP is the process of using algorithms to automatically identify and understand the emotions or opinions expressed in a piece of text, whether they are positive, negative, or neutral.


**Why use  sentiment analysis?**

Sentiment analysis in NLP is used to quickly grasp how people feel about something, whether it's a product, topic, or event. It helps businesses, researchers, and individuals understand public opinion, make informed decisions, and tailor their responses accordingly.

### Convert into base or root form of word

Converting words into their base or root form in NLP involves changing words to their simplest, canonical forms. This helps group similar words together, regardless of their tense or plural forms, making text analysis more accurate.This process, known as "lemmatization," aids in tasks like text classification and search by reducing words to their core meanings.

The line "nlp = spacy.load('en_core_web_sm')" loads a pre-trained language processing model called 'en_core_web_sm' from the SpaCy library for English language tasks. This model contains information about words, grammar, and meaning, enabling the program to understand and analyze text in English. Once loaded, you can use the 'nlp' object to perform various language processing tasks, like tokenization, part-of-speech tagging, and named entity recognition, making NLP tasks easier and more accurate.

```python
nlp = spacy.load('en_core_web_sm')
```

```python
x = 'this is chocolates.what is times ? this balls'
```

```python
def make_to_base(x):
    x = str(x)
    x_list = []
    doc = nlp(x)

    for token in doc:
        lemma = token.lemma_
        if lemma == '-PRON-' or lemma == 'be':
           lemma = token.text

        x_list.append(lemma)
    return ' '.join(x_list)
```

```python
make_to_base(x)
```

```python
text_df['text'] = text_df['text'].apply(lambda x: make_to_base(x))
```

```python
# Lists to store words based on sentiment
negative_words = []
positive_words = []
neutral_words = []

words = text.split()


# Perform sentiment analysis for each word
for word in words:
    blob = TextBlob(word)

    # blob.sentiment.polarity is a feature provided by TextBlob that quantifies the sentiment of a text.
    # It calculates a numerical value that indicates whether the text is positive, negative, or neutral based on the arrangement of words and their emotional impact.
    sentiment_score = blob.sentiment.polarity
    words = blob.words

    # Categorize words based on sentiment score
    if sentiment_score < 0:
        negative_words.extend(words)
    elif sentiment_score > 0:
        positive_words.extend(words)
    else:
        neutral_words.extend(words)

print('negative_words',negative_words)
print('positive_words',positive_words)
print('neutral_words',neutral_words)
```

CountVectorizer: This is a method used for text preprocessing in natural language processing (NLP). It converts a collection of text documents into a matrix of token counts, where each row represents a document, and each column represents a unique word or combination of words (depending on parameters like ngram_range)

.ngram_range=(1,2): This parameter specifies the range of n-grams to consider during vectorization. In this case, it considers both single words (unigrams) and pairs of consecutive words (bigrams). So, it creates features not only for individual words but also for combinations of two words occurring together.

.fit(sentiment_df_list['text']): This fits the CountVectorizer to the text data in the "text" column of the DataFrame or list sentiment_df_list. This step analyzes the text data to build the vocabulary (unique words or n-grams) and prepares the CountVectorizer for transforming text data into a matrix of token counts later on.

Precision: Precision measures the proportion of correctly predicted instances of a class among all instances predicted as that class. For example, for the "Neutral" class, 58% of the instances predicted as "Neutral" were actually "Neutral".

Recall: Recall, also known as sensitivity, measures the proportion of correctly predicted instances of a class among all actual instances of that class. For example, for the "Neutral" class, 97% of the actual "Neutral" instances were correctly predicted as "Neutral".

F1-score: The F1-score is the harmonic mean of precision and recall. It provides a balance between precision and recall. It's useful when there is an uneven class distribution.
Support: Support indicates the number of actual occurrences of each class in the test dataset.

Accuracy: Accuracy measures the proportion of correctly classified instances among all instances in the test dataset. In this case, the overall accuracy of the model is 60%.

Macro Avg: The macro average calculates the metrics' mean without considering class imbalance. It calculates the average precision, recall, and F1-score across all classes.

Weighted Avg: The weighted average calculates the metrics' mean while considering class imbalance. It calculates the average precision, recall, and F1-score across all classes, weighted by the number of instances of each class.
